```python
def tokenization_df(text):
    benc = xlmr_tokenizer(text)
    word_ids = benc.word_ids()
    token_ids = benc['input_ids']
    tokens = [xlmr_tokenizer.convert_ids_to_tokens(i) for i in token_ids]
    words = []
    for word_id in word_ids:
        if word_id is not None:
            start, end = benc.word_to_chars(word_id)
            words.append(text[start:end])
        else:
            words.append(None)
        
    df = pd.DataFrame({
        "token_ids": token_ids,
        "tokens": tokens,
        "word_ids": word_ids,
        "words": words
    })
    return df
    
```

```python
def clean_word(word: str) -> str:
    if not word:
        return word
    punc_end_or_beginning = re.compile(r"(^\W+|\W+$)")
    
    return punc_end_or_beginning.sub("", word.strip()).lower() or word
```

```python
def gather_segmentations(tok_df):
    out = defaultdict(list)
    for word, tok in zip(tok_df.words, tok_df.tokens):
        out[word].append(tok)
    return {
        word: tuple(toks)
        for word, toks in out.items()
    }
```

## Build the ultimate mapping

```python
token_word_joint = defaultdict(lambda: defaultdict(Counter))
token_marg = defaultdict(Counter)
word_marg = defaultdict(Counter)
Ns = Counter()

word_length_dist = []
```

```python
for gsid, row in tqdm(flores_df.iterrows(), total=flores_df.shape[0]):
    
    lang = row.language
    tok_df = tokenization_df(row.text)
    
    segms = gather_segmentations(tok_df)
    for word, subwords in segms.items():
        if word is not None:
            word_length_dist.append((lang, word, len(subwords)))
    
    for token, word in zip(tok_df.tokens, tok_df.words):
        if word is None:
            continue
        token_word_joint[lang][token][word] += 1
        word_marg[lang][word] += 1
        token_marg[lang][token] += 1
        Ns[lang] += 1
```

```python
word_given_token = defaultdict(lambda: defaultdict(lambda: defaultdict(float)))
token_given_word = defaultdict(lambda: defaultdict(lambda: defaultdict(float)))

for lang, dist in token_word_joint.items():
    for token, word_dist in dist.items():
        for word, joint_occurrence_count in word_dist.items():
            token_given_word[lang][word][token] = joint_occurrence_count / word_marg[lang][word]
            word_given_token[lang][token][word] = joint_occurrence_count / token_marg[lang][token]
```

```python
compression_analysis = pd.DataFrame(
    [
        (
            lang, 
            len(token_given_word[lang]), 
            len(word_given_token[lang])
        ) 
        for lang in token_word_joint
    ], columns=['language', 'n_words', 'n_subwords'])\
.assign(
    vocab_reduction_ratio=lambda df: (df.n_subwords/df.n_words).round(2)
)
```

```python
compression_analysis.vocab_reduction_ratio.plot(kind='hist', color='darkred', alpha=0.5, bins=30, title=f"Vocab reduction ratio\n({MODEL_NAME})")
```

```python
compression_analysis\
    .set_index('language')\
    .vocab_reduction_ratio\
    .sort_values()\
    .tail(30)\
    .plot(kind='barh', figsize=(6, 18), color='darkred', alpha=0.5, 
          title=f"Top 30 worst compressed languages\n({MODEL_NAME})")
plt.axvline(x=1, color='k', linestyle='--', alpha=0.5)
```

```python
import unimorph as um
```

```python
def analyze_with_unimorph(word, lang):
    lines = um.analyze_word(word, lang=lang).split("\n")
    analyses = [line.split("\t")[-1].split(";") for line in lines]
    return Counter(m for a in analyses for m in a if m)
```

```python
import itertools as it
```

```python
LANG="fin"

subword_to_msd = defaultdict(Counter)
subword_to_msd = defaultdict(Counter)
unimorph_fin = um.load_dataset(lang=LANG)
word_to_msd = defaultdict(Counter)

msd_to_word_data = defaultdict(Counter)

um_df = um.load_dataset(lang=LANG)

for word, feats in zip(um_df.form, um_df.features):
    for feat in feats.split(";"):
        word_to_msd[word][feat] += 1


for word in token_given_word[LANG]:
    msd_ctr = word_to_msd.get(word)
    if msd_ctr is not None:
        msds = list(msd_ctr)
        subwords = [sw for sw in token_given_word[LANG][word]]
        for morph, subword in it.product(msds, subwords):
            subword_to_msd[subword][morph] += 1
            msd_to_word_data[morph][word] += 1
            
df_subword_to_msd = pd.DataFrame(subword_to_msd).T.fillna(0)

_ = df_subword_to_msd\
    .assign(num_msd_features_on=lambda df: df.sum(axis=1))\
    .assign(token_length=lambda df: [len(sw) for sw in df.index.to_list()])\
    .sort_values("num_msd_features_on", ascending=False)

# _['num_msd_features_on_standardized'] = (_.num_msd_features_on - _.num_msd_features_on.mean()) / _.num_msd_features_on.std()
```
